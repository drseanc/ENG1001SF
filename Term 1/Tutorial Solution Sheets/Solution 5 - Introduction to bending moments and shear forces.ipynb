### Question 1

**(i) Derive expressions for the shear force and bending moment as a function of position along the beam, x.** 

**(ii) Sketch the shear force and bending moment diagram for the beam.** 

**(iii) Stretch goal - try to write some Python code to plot the shear force and bending moment diagrams**

**Hint:** You will need to generate a range of x-values and use a **for loop** to cycle through them. Within the for loop, use **if/elif** conditions to test which of your equations should be used to calculate shear force and bending moment for that value of x.

### Answer 1 (i)

For $0\leq x \leq 4$:

\begin{align}
&M(x) = 54.69x\\ 
&S(x) = 54.69 
\end{align}

For $4\leq x \leq 8$:

\begin{align}
&M(x) = 4.69x + 200\\ 
&S(x) = 4.69 
\end{align}

For $8\leq x \leq 13$:

\begin{align}
&M(x) = -5x^2 + 84.69x -120\\ 
&S(x) = -10x + 84.69
\end{align}

For $13\leq x \leq 16$:

\begin{align}
&M(x) = -45.31x+725\\ 
&S(x) = -45.31
\end{align}

### Answer 1 (iii)

```python
delta_x = 0.01 #Increment for values of x
X = np.arange(0,16+delta_x,delta_x) #Generate a range of x values
SF = np.array([]) #Initialise an array to hold shear force values at each x position
BM = np.array([]) #Initialise an array to hold bending moment values at each x position

#Loop through all values of x in X
for x in X:
    #Check to see which set of equations apply then calculate the shear force and bending moment
    if(x>=0 and x<=4):
        M = 54.69*x
        S = 54.69
    elif(x>4 and x<=8):
        M = 4.69*x+200
        S = 4.69
    elif(x>8 and x<=13):
        M = -5*x**2 +84.69*x-120
        S = -10*x+84.69
    elif(x>13 and x<=16):
        M = -45.31*x + 725
        S = -45.31
           
    #Store the valuer of shear force and bending moment in the arrays initialised above
    BM = np.append(BM, M)
    SF = np.append(SF, S)

#Plot SFD    
fig = plt.figure() #Define a new figure
axes = fig.add_axes([0.1,0.1,2,1]) #Add a set of axes to the figure defining axes location and size
axes.plot(X,SF,) #Plot x versus M with a blue line with circles indicated data points
axes.set_xlim(0,16) 
axes.set_xlabel('Distance (m)') #set an x-label
axes.set_ylabel('Shear Force (kN)') #set a y-label
axes.set_title('Shear Force Diagram') #set a title
axes.grid() #Show a grid

#Plot BMD
fig = plt.figure() #Define a new figure
axes = fig.add_axes([0.1,0.1,2,1]) #Add a set of axes to the figure defining axes location and size
axes.plot(X,BM,) #Plot x versus M with a blue line with circles indicated data points
axes.set_ylim(250,0) #Invert y-axis (for BMD positive downwards)
axes.set_xlim(0,16) 
axes.set_xlabel('Distance (m)') #set an x-label
axes.set_ylabel('Bending moment (kNm)') #set a y-label
axes.set_title('Bending Moment Diagram') #set a title
axes.grid() #Show a grid
```

### Answer 2 (iii)

```python
delta_x = 0.01 #Increment for values of x
X = np.arange(0,18+delta_x,delta_x) #Generate a range of x values
SF = np.array([]) #Initialise an array to hold shear force values at each x position
BM = np.array([]) #Initialise an array to hold bending moment values at each x position

#Loop through all values of x in X
for x in X:
    #Check to see which set of equations apply then calculate the shear force and bending moment
    if(x>=0 and x<=3):
        M = -(50*x**3)/24
        S = -(50*x**2)/8
    elif(x>3 and x<=4):
        M = -(50*x**3)/24 + 187.75*x - 563.25
        S = -(50*x**2)/8 + 187.75
    elif(x>4 and x<=9):
        M = -10*x**2 + 167.75*x - 456.25
        S = -20*x + 167.75
    elif(x>9 and x<=15):
        M = -10*x**2 + 167.75*x - 356.25
        S = -20*x + 167.75
    elif(x>15 and x<=18):
        M = -10*x**2 + 360*x - 3240
        S = -20*x + 360
           
    #Store the valuer of shear force and bending moment in the arrays initialised above
    BM = np.append(BM, M)
    SF = np.append(SF, S)

#Plot SFD    
fig = plt.figure() #Define a new figure
axes = fig.add_axes([0.1,0.1,2,1]) #Add a set of axes to the figure defining axes location and size
axes.plot(X,SF,) #Plot x versus M with a blue line with circles indicated data points
axes.set_xlim(0,18) 
axes.set_xlabel('Distance (m)') #set an x-label
axes.set_ylabel('Shear Force (kN)') #set a y-label
axes.set_title('Shear Force Diagram') #set a title
axes.grid() #Show a grid

#Plot BMD
fig = plt.figure() #Define a new figure
axes = fig.add_axes([0.1,0.1,2,1]) #Add a set of axes to the figure defining axes location and size
axes.plot(X,BM,) #Plot x versus M with a blue line with circles indicated data points
axes.set_ylim(350,-100) #Invert y-axis (for BMD positive downwards)
axes.set_xlim(0,18) 
axes.set_xlabel('Distance (m)') #set an x-label
axes.set_ylabel('Bending moment (kNm)') #set a y-label
axes.set_title('Bending Moment Diagram') #set a title
axes.grid() #Show a grid
```

(iii) Max moment between A and B:

M_{\text{max}} = \frac{9wL^2}{128}
