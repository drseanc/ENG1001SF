# 1.0 Moment of Inertia
(a.k.a. second moment of area)

Consider some arbitrary area with centroid $C$ with respect to the $x$ and $y$ axis. Now imagine a small element of area $\mathrm{d}A$ located at position $[x,y]$. 

By definition the second moment of area for this arbitrary area about the $x$ and $y$ axis is,

\begin{equation}
I_x = \int y^2 \mathrm{d}A
\end{equation}

\begin{equation}
I_y = \int x^2 \mathrm{d}A
\end{equation}

Thus the second moment of area about an axis is equivalent to multiplying each sub-element of area by the square of its distance from the axis. The integration sign here simply indicates that we do this for all such sub-elements of area that make up the complete area. 

## 1.1 Performing the integration

We can perform an analytical integration for any area for which $x$ and $y$ can be expressed as a function. For example, consider the rectangular area with centroid $C$ located at the origin of an $x-y$ axis system.

In this case, the element of area $\mathrm{d}A$ is selected to be a horizontal slice across the full with of the section located a distance $y$ from the x-axis. Therefore we can say,

\begin{equation}
\mathrm{d}A = b\:\mathrm{d}y
\end{equation}

Our expression for $I_x$ therefore becomes,

\begin{equation}
I_x = \int y^2 \mathrm{d}A = \int_{-h/2}^{h/2}y^2b\mathrm{d}y
\end{equation}

Evaluating this integration yields,

\begin{equation}
\boxed{
I_x = \frac{bh^3}{12}
}
\end{equation}

We can perform the same procedure with a vertical strip of area to determine the second moment of area with respect to the y-axis, yielding,

\begin{equation}
\boxed{
I_y = \frac{hb^3}{12}
}
\end{equation}

The same fundamental procedure can be used to determine $I_x$ and $I_y$ for any cross-sectional area. In the case above, the axes were a set of centroidal axes because they passed through the centroid of the area. This is convenient, however we can determine the second moment of area about any parallel set of axes by using the parallel axis theorem we've seen previously. 

# 2.0 Polar Moment of Inertia

In the preceding section, $I_x$ and $I_y$ were calculated about a set of axis lying on the same plane as the area. If we consider a moment of inertia about an axis **perpendicular** to the area, this is referred to as the **Polar moment of Inertia**, $I_p$. This is why it crops up when dealing with torsion, because the axis about which twisting is taking place is perpendicular to the cross-sectional area under consideration. 

The polar moment of inertia is defined as,

\begin{equation}
I_p = \int \rho^2\mathrm{d}A
\end{equation}

where $\rho$ is the distance between an element of area $\mathrm{d}A$ and the point $O$ about which the axis passes(see below).

Note from the image above that $\rho^2 = x^2 + y^2$. From this observation we obtain an important result,

\begin{equation}
I_p = \int \rho^2\mathrm{d}A = \int (x^2+y^2)\mathrm{d}A = \int x^2\mathrm{d}A + \int y^2 \mathrm{d}A
\end{equation}

\begin{equation}
\boxed{
I_p = I_x + I_y
}
\end{equation}

The polar moment of inertia can be obtained as the sum of the moments of inertia with respect to any two perpendicular axis passing through the same point and lying in the plane of the area.

## 2.1 Performing the integration

To demonstrate calculating the polar moment of inertia for a circle about a centroidal axis, consider the area below and the ring of area $\mathrm{d}A$, located at radius $\rho$ with thickness $\mathrm{d}\rho$.

The integration required is,

\begin{equation}
I_p = \int \rho^2\mathrm{d}A = \int_0^r \rho^2\:2\pi\rho\:\mathrm{d}\rho
\end{equation}

\begin{equation}
\boxed{
I_p = \frac{\pi r^4}{2}
}
\end{equation}

Now if we substitute $r=d/2$ where $d$ is the diameter of the circle, we obtain,

\begin{equation}
\boxed{
I_p = \frac{\pi d^4}{32}
}
\end{equation}
