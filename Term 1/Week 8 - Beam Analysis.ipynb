## 1.0 Rotational Hinges / Pins 

We've already encountered pin joints in our study of truss structures in week 3. Remember, these joints only allow force transmission, no moment can be transferred across a pin joint. Pin joints can also appear in beam and frame structures. 

When analysing beam structures previously, the first step was to determine the support reactions. We established in week 3 that we could only determine up to three unknown reactions using our three equations of statics. 

However, if our structure contains a rotational pin, this limit is increased by 1 to 4. Therefore we can calculate 4 unknown reactions and subsequently determine the shear and moment diagrams. This week learn how to analyse beams with pins. 

This week's material assumes you have a good grasp of the material we've already covered on the topic of shear force and bending moment diagrams. Hopefully this is the case following your review week - in which case the introduction of rotational pins presents only a minor complication.

Let's start by considering the propped cantilever below.


<center>Fig 1. Statically indeterminate propped cantilever</center>

As we've already seen, this structure is statically indeterminate to $1^{\circ}$. However, now consider the same structure with a rotational hinge/pin introduced between C and D at E.


<center>Fig 2. Statically determinate propped cantilever with internal hinge</center>

The presence of the hinge reduced the degree of indeterminacy by 1 and since the structure was only statically indeterminate to $1^{\circ}$, the hinge renders it statically determinate. Now the question is how do we capitalise on the hinge and determine the support reactions?

Whenever we encounter a hinge like this, our strategy is to **break the structure at the hinge** and consider the sub-structures on either side of the hinge separately.


<center>Fig 3. Statically determinate propped cantilever with internal hinge, separated at the hinge</center>

Note that we now denote the locations immediately to the left and right of the hinge as $E_L$ and $E_R$ respectively. By dividing or cutting the structure at the hinge, we reveal the force that is transmitted through the hinge. $F_{E,R}$ is the force exerted on the right side sub-structure by the left sub-structure. Similarly, $F_{E,L}$ is the equal and opposite force exerted on the left side sub-structure by the right sub-structure. Although we appear to be introducing two forces ($F_{E,L}$ & $F_{E,R}$), both are in fact just different representations of the same force, the force transmitted through the hinge. 

Now, the left sub-structure still contains 4 unknown forces (3 reactions + $F_{E,L}$. However, the right sub-structure has only 2 unknowns and can therefore be analysed. 

Taking the sum of the moments about point B,

\begin{equation*}
\sum M_{\text{B}} = 0 \:\:(\curvearrowright +ve)
\end{equation*}

\begin{equation*}
F_{E,R} \times 7.5 - 30\times 6 = 0
\end{equation*}

\begin{equation*}
\therefore F_{E,R} = 24\:\text{kN}\:\:\text{(acting upwards as shown)}
\end{equation*}

Now evaluating the sum of the vertical forces for the right sub-structure yields the vertical reaction at B, $V_B$,

\begin{equation*}
\sum F_y = 0\:\:(\uparrow+ve)
\end{equation*}

\begin{equation*}
V_B = 6 \:\:\text{kN (upwards)}
\end{equation*}

Now we can also state the force on the left of the hinge, 

\begin{equation*}
\therefore F_{E,L} = 24\:\text{kN}\:\:\text{(acting downwards as shown)}
\end{equation*}

This renders the left sub-structure statically determinate and open for analysis. Taking moments about point A,

\begin{equation*}
\sum M_{\text{A}} = 0 \:\:(\curvearrowright +ve)
\end{equation*}

\begin{equation*}
M_A = (45\times 3) + (24\times 7.5) = 315\:\:\text{kNm (counter-clockwise)}
\end{equation*}

By inspection we can see that the horizontal reaction at A, $H_A = 0$ and the vertical reaction $V_A = 69\:\:\text{kN}$ upwards. We can summarise our findings so far in a free body diagram below. Note that both force and moment equilibrium for the complete structure are maintained. 


<center>Fig 4. Free body diagram showing reactions</center>

We can now evaluate the shear force and bending moment diagram for the structure utilising what we've learned over the previous two weeks.

As you can see from the preceding example, the introduction of a rotational hinge introduces only one extra step into the analysis you're already familiar with from previous weeks. 

The extra step is to break the structure at the hinge(s) in order to allow the reactions to be identified. The analysis to determine the shear force and bending moment diagram continues in the normal way thereafter. 

One final point to remember; since there can be no bending moment transmitted through a rotational hinge, the bending moment diagram will always show zero moment at the location of a hinge.
