## 1.0 Introduction

In our study of mechanics so far we have considered static problems, i.e. the velocity (rate of change of position) of the objects under consideration was equal to zero. We now embark on the study of dynamics, i.e. the behaviour of objects whose velocity is non-zero and in many cases is either increasing or decreasing at a particular rate or acceleration.

This week we will focus on describing and analysing motion in a straight line and in doing so we will formalise the concepts of position, velocity and acceleration. The ability to mathematically model and therefore predict straight-line (or *translational*) motion has an enormous number of practical applications from navigation to the mechanism in a 3D printer to the tiny Taptic Engine in your smartphone.

## 2.0 Position, Velocity and Acceleration

### 2.1 Position

We typically describe position as a vector in 2D or 3D space (alternatives will be discussed later). This vector exists within a coordinate reference frame, with an origin point $O$, see below. 

![title](img/T2W4/img1.jpg)
<center>Position vector $\mathbf{r}$ in a 3-dimensional reference frame with origin $O$.</center>

The position of some point $P$ can be described by the vector $\mathbf{r}$ (note the bold font for denoting a vector quantity) which is simply the coordinates of the object in the given reference frame. If the object is moving then we can say that $\mathbf{r}$ is a function of time, $\mathbf{r}=\mathbf{r}(t)$.

### 2.2 Velocity

Velocity is the rate of change of position. The velocity $\mathbf{v}$ of some point $P$ in a given reference frame is defined as,

\begin{equation}
v=\frac{\mathrm{d}\mathbf{r}}{\mathrm{d}t} = \lim_{\Delta t\rightarrow 0} \frac{\mathbf{r}(t+\Delta T)-\mathbf{r}(t)}{\Delta t} = \frac{\text{change in position}}{\text{time interval}}
\end{equation}

Remember, $\mathbf{r}$ was the vector describing the position of $P$. We can visualise the change in position (numerator above) as a vector, 

![title](img/T2W4/img2.jpg)
<center>Change in position as a vector quantity.</center>

### 2.3 Acceleration

Acceleration is the rate of change of velocity. Therefore, the acceleration $\mathbf{a}$ of the point $P$ in a given reference frame is defined as,

\begin{equation}
a=\frac{\mathrm{d}\mathbf{v}}{\mathrm{d}t} = \lim_{\Delta t\rightarrow 0} \frac{\mathbf{v}(t+\Delta T)-\mathbf{v}(t)}{\Delta t} = \frac{\text{change in velocity}}{\text{time interval}}
\end{equation}

From this we note that acceleration can also be obtained as the second differential of position with respect to time,

\begin{equation}
\mathbf{a} = \frac{\mathrm{d}^2\mathbf{r}}{\mathrm{d}t^2}
\end{equation}

We can again visualise the vector representation of the change in velocity.

![title](img/T2W4/img3.jpg)
<center>Change in velocity as a vector quantity.</center>

**Scalars and Vectors**

It is important to realise that position, velocity and acceleration are all vector quantities. By definition this means they have a magnitude and direction. Scalars have magnitude only, e.g. speed is a scalar quantity while velocity is clearly a vector. 

The derivative of a vector shares some of the properties of the
derivative of a scalar function. *The derivative of the sum of two vector functions* $\mathbf{u}$ and $\mathbf{w}$ i.e. $\mathbf{u}+\mathbf{w}$ is,

\begin{equation}
\frac{\mathrm{d}}{\mathrm{d}t}(\mathbf{u}+\mathbf{w}) = \frac{\mathrm{d}\mathbf{u}}{\mathrm{d}t} + \frac{\mathrm{d}\mathbf{w}}{\mathrm{d}t}
\end{equation}

*The derivative of the product of a scalar function $f$ and a vector $\mathbf{u}$, i.e. $f\mathbf{u}$ is,*

\begin{equation}
\frac{\mathrm{d}}{\mathrm{d}t}(f\mathbf{u}) = \frac{\mathrm{d}f}{\mathrm{d}t}\mathbf{u} + \frac{\mathrm{d}\mathbf{u}}{\mathrm{d}t}f
\end{equation}

## 3.0 Describing Motion

Having formalised the definition of position, velocity and acceleration above, we now focus on applying these mathematical definitions to the description of linear motion. Consider a point $P$ on a straight line some distance $s$ from an origin point $O$ (below). The displacement during the time interval between $t_0$ and $t$ is the change in position $s(t) - s(t_0)$.

![title](img/T2W4/img4.jpg)
<center>Point $P$ undergoing a displacement $s$ between time $t_0$ and $t$.</center>

By introducing a **unit vector** $\mathbf{e}$ parallel to the line and pointing from $O$ to $P$, we can write the position vector $\mathbf{r}$ as,

\begin{equation}
\mathbf{r}=s\mathbf{e}
\end{equation}

where $\mathbf{r}$ is the vector describing the position of $P$ relative to $O$ and $s$ is a scalar.

Using equation (5) above, we can obtain the velocity of $P$ relative to $O$ as,

\begin{equation}
\mathbf{v}=\frac{\mathrm{d}\mathbf{r}}{\mathrm{d}t} = \frac{\mathrm{d}}{\mathrm{d}t}s\mathbf{e} = \frac{\mathrm{d}s}{\mathrm{d}t}\mathbf{e} + \frac{\mathrm{d}\mathbf{e}}{\mathrm{d}t}s
\end{equation}

Since the magnitude and direction of the unit vector $\mathbf{e}$ is constant, 

\begin{equation}
\frac{\mathrm{d}\mathbf{e}}{\mathrm{d}t} = 0
\end{equation}

$\mathbf{v}$ therefore simplifies to,

\begin{equation}
\boxed{
\mathbf{v}=\frac{\mathrm{d}s}{\mathrm{d}t}\mathbf{e}
}
\end{equation}

or 

\begin{equation}
\mathbf{v} = v\mathbf{e}
\end{equation}

where,

\begin{equation}
v = \frac{\mathrm{d}s}{\mathrm{d}t}
\end{equation}

We note that if we graph $s$ against $t$ (below), $v$ is the slope of a tangent to the graph at time $t$. 

![title](img/T2W4/img5.jpg)
<center>Graph of $s$ versus $t$ showing slope $v$ at time $t$.</center>

We can similarly obtain an expression for the acceleration of $P$ relative to $O$ as,

\begin{equation}
\mathbf{a} = \frac{\mathrm{d}\mathbf{v}}{\mathrm{d}t} = \frac{\mathrm{d}}{\mathrm{d}t}v\mathbf{e} 
\end{equation}

\begin{equation}
\boxed{
\mathbf{a} = \frac{\mathrm{d}v}{\mathrm{d}t}\mathbf{e}
}
\end{equation}

or

\begin{equation}
\mathbf{a} = a\mathbf{e}
\end{equation}

where,

\begin{equation}
a=\frac{\mathrm{d}v}{\mathrm{d}t} = \frac{\mathrm{d}^2s}{\mathrm{d}t^2}
\end{equation}



Again note that $a$ is the slope of the tangent to the curve of $v$ against $t$ at time $t$.

![title](img/T2W4/img6.jpg)
<center>Graph of $v$ versus $t$ showing slope $a$ at time $t$.</center>

### Example 1

![title](img/T2W4/img7.jpg)

A speedboat's position during the time interval $2\leq t \leq 10$ is given by,

\begin{equation*}
s = 4t + 1.6t^2 - 0.08t^3\:m
\end{equation*}

1. Determine the speedboat's velocity and acceleration at $t=4\:s$


2. What is the speedboat’s maximum velocity during this interval of time and when does it occur?

**Solution**

(a) We can obtain expressions for velocity and acceleration magnitude by differentiating,

\begin{align*}
v &= \frac{\mathrm{d}s}{\mathrm{d}t} = 4 + 3.2t-0.24t^2\\
a &= \frac{\mathrm{d}v}{\mathrm{d}t} = 3.2-0.48t\\
\end{align*}

These can then be evaluated at $t=4\:s$.

```python
t = 4 #(s)
v = 4 + 3.2*t - 0.24*t**2 #(m/s)
a = 3.2-0.48*t #(m/s^2)
print('v(4) = {one} m/s'.format(one=round(v,2)))
print('a(4) = {one} m/s^2'.format(one=round(a,2)))
```

(b) Remember that these are simple functions and so the usual techniques we've used previously can be used to identify the location of local maxima and mimima. 

The maximum velocity occurs when the tangent to the curve $v(t)$ is zero. In other words, when $a = 0$. Therefore letting $a=0$ and solving for $t$ yields,

\begin{align*}
3.2-0.48t &= 0\\
t &=6.67\:s
\end{align*}

We can now evaluate $v(t=6.67)$ to identify the (maximum) velocity at this time.

```python
t = 6.67 #(s)
v = 4 + 3.2*t - 0.24*t**2 #(m/s)
print('v(t=6.67) = {one} m/s'.format(one=round(v,2)))
```

### 3.1 Acceleration Specified as a function of time

In the previous example we started with a function describing position as a function of time. From this we obtained position and velocity by differentiating. Alternatively, if we know the acceleration as a function of time (starting from the opposite end of the problem), we can obtain velocity and position through integration,

\begin{equation}
\frac{\mathrm{d}v}{\mathrm{d}t} = a(t)
\end{equation}

\begin{equation}
v = \int a(t) \mathrm{d}t + A
\end{equation}

where $A$ is an unknown constant of integration. If we integrate again, we obtain the position $s$,

\begin{equation}
s = \int v(t) \mathrm{d}t +B
\end{equation}

The unknown constants of integration can be determined if we know the boundary conditions for the system, i.e. the position and velocity at some instant of time. An alternative approach which ultimately yields the same result is to integrate in terms of definite integrals, rearranging equation (16),

\begin{equation}
\mathrm{d}v = a(t)\mathrm{d}t
\end{equation}

Integrating both sides between the limits specified,

\begin{equation}
\int_{v_0}^v\mathrm{d}v = \int_{t_o}^t a\mathrm{d}t
\end{equation}

\begin{equation}
\boxed{
v = v_0 + \int_{t_o}^t a\mathrm{d}t
}
\end{equation}

Note that this still requires us to know the initial velocity, $v_0$. Applying the same definite integral approach to the integration of $v$ yields,

\begin{equation}
\boxed{
s = s_0 + \int_{t_o}^t v\mathrm{d}t
}
\end{equation}



### 3.2 Constant Acceleration

If the acceleration is constant, $a=a_0$, the previous integrations become,

\begin{equation}
v=v_0 + \int_{t_o}^t a_0\mathrm{d}t
\end{equation}

\begin{equation}
\boxed{
v=v_0 + a_0(t-t_0)
}
\end{equation}

If we now substitute equation (24) into (22),

\begin{equation}
s=s_0 + \int_{t_o}^t \left[v_0 + a_0(t-t_0) \right] \mathrm{d}t
\end{equation}

we obtain the equation for position of an object subject to a constant acceleration $a_0$,

\begin{equation}
\boxed{
s = s_0 + v_0(t-t_0) + \frac{1}{2}a_0(t-t_0)^2
}
\end{equation}

### 3.3 Constant Acceleration with Velocity as a function of Position

We now consider the case when the acceleration is still constant, but the velocity is a function of position, $v = v(s)$. We know that acceleration is the first derivative of velocity,

\begin{equation}
a_0 = \frac{\mathrm{d}v}{\mathrm{d}t}
\end{equation}

Since,

\begin{equation}
\frac{\mathrm{d}v}{\mathrm{d}t} = \frac{\mathrm{d}v}{\mathrm{d}s}\frac{\mathrm{d}s}{\mathrm{d}t} = \frac{\mathrm{d}v}{\mathrm{d}s}v
\end{equation}

we can say,

\begin{equation}
a_0 = \frac{\mathrm{d}v}{\mathrm{d}s}v
\end{equation}

Rearranging the equation and integrating,

\begin{equation}
\int_{v_0}^v v\mathrm{d}v = \int_{s_0}^s a_0\mathrm{d}s
\end{equation}

which yields,

\begin{equation}
\boxed{
v^2 = v_0^2 + 2a_0(s-s_0)
}
\end{equation}

Equations (24), (26) and (31) are helpful and commonly deployed to model straight-line motion, however **remember that they can only be used when the acceleration is constant.**

### 3.4 Acceleration as a function of velocity <mark>(if you know $a=a(v)$)</mark>


Aerodynamic and hydrodynamic forces can cause an object's acceleration to depend on its velocity. For example, consider a bullet or rocket travelling at high speed through air. As it moves through the air the drag force it experiences, $F_D$ is a function of its velocity magnitude, $v$,

\begin{equation}
F_D = \frac{1}{2}\rho v^2C_D A
\end{equation}

where $A$ is the cross-sectional area, $C_D$ is the drag coefficient and $\rho$ is the density of the air. The key point here is that one of the forces governing the acceleration of the body varies with velocity, therefore the acceleration will vary with velocity. 

Another common example of a velocity proportional force is the force provided by an oil-filled damper within a shock-absorber. If we try to compress a shock absorber, the damper will exert a resisting force proportional to the speed with which we try to compress it. 

Suppose that the acceleration is a known function of velocity, 

\begin{equation}
a(v) = \frac{\mathrm{d}v}{\mathrm{d}t}
\end{equation}

In order to determine the velocity, we cannot integrate this equation with respect to time because $a(v)$ is not known as a function of time. But we can separate the variables,

\begin{equation}
\frac{\mathrm{d}v}{a(v)} = \mathrm{d}t
\end{equation}

and then integrate,

\begin{equation}
\int_{v_0}^v \frac{\mathrm{d}v}{a(v)} = \int_{t_0}^t \mathrm{d}t
\end{equation}

This integration can be solved to determine the velocity as a function of time. If position is required, simply integrate the equation just obtained for $v(t)$ to determine $s(t)$.

Starting with the acceleration as a function of velocity, $a(v)$ we can also determine the *velocity as a function of position* using the chain rule, from equation (28),

\begin{equation}
a(v) = \frac{\mathrm{d}v}{\mathrm{d}t} = \frac{\mathrm{d}v}{\mathrm{d}s}\frac{\mathrm{d}s}{\mathrm{d}t} = \frac{\mathrm{d}v}{\mathrm{d}s}v
\end{equation}

Separating variables and integrating yields,

\begin{equation}
\int_{v_0}^v\frac{v\mathrm{d}v}{a(v)} = \int_{s_0}^s\mathrm{d}s
\end{equation}

The solution to this integration will yield an expression for velocity as a function of position, based on an original expression for acceleration as a function of velocity.

### 3.5 Acceleration as a function of position <mark>(if you know $a=a(s)$)</mark>


One of the simplest examples of a system subject to a force that depends on position is a spring. The force in a spring with spring constant $k$ is,

\begin{equation}
F_{\text{spring}}=k\times \delta s
\end{equation}

where $\delta s$ is the extension of the spring. As a result, the acceleration of the spring (or more usually mass attached to the spring) will be a function of position $a=a(s)$. We therefore have,

\begin{equation}
a(s) = \frac{\mathrm{d}v}{\mathrm{d}t}
\end{equation}

Using the chain rule again,

\begin{equation}
a(s) = \frac{\mathrm{d}v}{\mathrm{d}t} = \frac{\mathrm{d}v}{\mathrm{d}s}\frac{\mathrm{d}s}{\mathrm{d}t} = \frac{\mathrm{d}v}{\mathrm{d}s}v
\end{equation}

Separating variables and integrating,

\begin{equation}
\int_{v_0}^v v\mathrm{d}v = \int_{s_0}^s a(s) \mathrm{d}s
\end{equation}

This integration can be evaluated to determine $v(s)$. Now recall from equation (11) that $v(s)$ is the first differential of $s$ with respect to $t$, we can again separate the variables and integrate, 

\begin{equation}
\int_{s_0}^s \frac{\mathrm{d}s}{v(s)} = \int_{t_0}^t \mathrm{d}t
\end{equation}

This integration will yield an expression for position as a function of time $s(t)$, from an initial expression for the acceleration as a function of position. 

### Example 4

After deploying a drag parachute, an aircraft has an acceleration $a(v)=-0.004v^2 \:m/s^2$.

- (a) Determine the time required for the velocity to decrease from $80\:m/s$ to $10\:m/s$.
- (b) What distance does the aircraft cover during this time?

**Solution**

(a) We can see that the acceleration is a function of the velocity, $a=a(v)$,

\begin{equation*}
a(v) = \frac{\mathrm{d}v}{\mathrm{d}t} = -0.004v^2
\end{equation*}

Using equation (35) we have,

\begin{equation*}
\int_{80}^v\frac{\mathrm{d}v}{v^2} = \int_0^t -0.004\mathrm{d}t
\end{equation*}

Now performing the integration yields,

\begin{equation*}
\left(-\frac{1}{v}+\frac{1}{80}\right) = -0.004t
\end{equation*}

Solving for $t$,

\begin{equation}
t = 250\left(\frac{1}{v} - \frac{1}{80}\right)
\end{equation}

From this equation we can readily find the time required for the aircraft to slow to $v=10\:m/s$.

```python
v0 = 80 #(m/s)
v_target = 10 #(m/s) 
t1 = 250*((1/v_target)-(1/v0))
print('a. The time required for the aircraft to slow to {A} m/s is {B} seconds.'.format(A=v_target, B=round(t1,1)))
```

We can confirm this visually by plotting $v$ as a function of $t$, rearrange equation (43),

\begin{equation}
v(t) = \frac{1}{\left(\frac{t}{250} + \frac{1}{80}\right)}
\end{equation}

```python
delT = 0.1 #(sec) Timestep
tMax = 30 #(sec) Max time to plot
t = np.arange(0,tMax+delT,delT) #(sec) Generate amn array of time values
v = 1/((t/250)+(1/80)) #(m/s) Evaluate the velocity at each time

#PLOTTING CODE
fig = plt.figure() #Define a new figure
axes = fig.add_axes([0.1,0.1,2,1]) #Add a set of axes to the figure defining axes location and size
axes.plot(t,v) #Plot t versus v
axes.plot([0,tMax],[10,10],'r',lw=0.75) #Plot a horizontal line marking the target velocity
axes.plot([t1,t1],[0,v0],'r',lw=0.75) #Plot a vertical red line indicating the calculated time to reach target velocity
axes.set_xlabel('time (sec)') #set an x-label
axes.set_ylabel('velocity (m/s)') #set a y-label
axes.set_title('Velocity-time history') #set a title
axes.set_xlim([0,tMax])
axes.set_ylim([0,v0])
axes.grid() #Show a grid
```

(b) From equation (36) we have,

\begin{equation*}
a(v) = \frac{\mathrm{d}v}{\mathrm{d}s}v =-0.004v^2
\end{equation*}

Separating variables,

\begin{equation*}
\frac{\mathrm{d}v}{v} = -0.004\mathrm{d}s
\end{equation*}

We can now integrate both sides, letting $v_0=80\:m/s$ when $s_0=0$,

\begin{equation*}
\int_{80}^v \frac{\mathrm{d}v}{v} = \int_0^s -0.004\mathrm{d}s
\end{equation*}

After performing the integration between the limits, we obtain,

\begin{equation*}
\ln v - \ln 80 = -0.004s
\end{equation*}

Which gives us the velocity as a function of position, given a known acceleration $a(v)$. Now solving for $s$,

\begin{equation*}
s=250\ln\left(\frac{80}{v}\right)
\end{equation*}

Therefore the distance required to slow to a velocity of $v=10\:m/s$ can be readily obtained.

```python
v0 = 80 #(m/s) Initial velocity
v_target = 10 #(m/s) Target velocity
s = 250*math.log(v0/v_target) #(m) Distance required to slow from initial to target velocity
print('b. The distance required for the aircraft to slow from {A} m/s to {B} m/s is {C} m.'.format(A=v0, B=v_target, C=round(s,1)))
```
